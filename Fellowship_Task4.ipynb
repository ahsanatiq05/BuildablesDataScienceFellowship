## Question 10

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = df.drop(columns = ['ID', 'Age', 'Country', 'Gender', 'Occupation'])
```

```python
df.head()
```

```python
ml_df = pd.get_dummies(df, columns=['Sleep_Quality', 'Stress_Level', 'Health_Issues'])
ml_df.head()
```

```python
X = ml_df.drop(columns = 'Heart_Rate')
y = ml_df['Heart_Rate']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
history = model.predict(X_test)
```

```python
hit_df = pd.DataFrame(history)
```

```python
hit_df.head()
```

```python
model.coef_
```
