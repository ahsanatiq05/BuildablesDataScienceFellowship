```python
data = sns.load_dataset('car_crashes')
```

```python
data = sns.load_dataset('car_crashes')
w = data.total
x = data.speeding
y = data.alcohol
z = data.ins_premium

plt.figure(figsize=(6, 6))
fig, axes = plt.subplots(2,2)
axes[0,0].hist(w, bins = 20)
axes[0,0].set_title("Total Frequency")

axes[0,1].scatter(y, w)
axes[0,1].set_title("Alcohol vs Total")

axes[1,0].boxplot(z)
axes[1,0].set_title("Loss Range")

axes[1,1].scatter(y, x)
axes[1,1].set_title("Alcohol vs Speeding")

plt.tight_layout()
plt.show()
```

## Question 18

```python

data_prev = pd.DataFrame({
    "hours": [1, 2, 3, 4, 5, 6, 7, 8, 9],
    "score": [20, 35, 50, 55, 65, 70, 80, 85, 90]
})


sns.set_style("whitegrid")  # options: white, dark, whitegrid, darkgrid, ticks
sns.set_palette("coolwarm")  # try "viridis", "pastel", "deep", "muted", "Set2"

plt.figure(figsize=(8,5))
sns.lineplot(data=data_prev, x="hours", y="score", marker="o")


plt.title("Study Hours vs Exam Score", fontsize=16, fontweight="bold")
plt.xlabel("Hours Studied", fontsize=12)
plt.ylabel("Score Achieved", fontsize=12)

plt.show()
```

## Question 19

```python
data = pd.DataFrame({
    "groups": ["Alif"] * 5 + ["Bay"] * 5,
    "score": [20, 35, 50, 55, 65, 70, 80, 85, 90, 90]
})
plt.figure(figsize = (8, 5))
sns.violinplot(x = data.groups, y = data.score, palette = "Set2")
plt.title("Score Distribution Across Groups")
plt.xlabel("Group")
plt.ylabel("Score")
plt.show()
```

```python


# Generate some data

data = np.random.randint(0, 100, size=200)

plt.figure(figsize=(8,5))

plt.hist(data, bins=20, alpha=0.6, density=True, label="Histogram")

sns.kdeplot(data, color="red", linewidth=2, label="KDE Curve")

plt.title("Hybrid Visualization: Histogram + KDE")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
